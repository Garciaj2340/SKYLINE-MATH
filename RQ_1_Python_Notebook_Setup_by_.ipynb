```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head()
```

:\Codebook 1 Screen Shot 2023-04-30 at 10.52.43 PM.png

```python
#Manage data for variable 1 
df['getting_up']= df['H4SP6']
df['getting_up']=df['getting_up'].loc[ df['getting_up']!=' ' ]
df['getting_up']=df['getting_up'].loc[ df['getting_up']!='6' ]
df['getting_up']=df['getting_up'].loc[ df['getting_up']!='8' ]
df['getting_up']=df['getting_up'].loc[ df['getting_up']!='5' ]
df['getting_up'].value_counts()
```

```python
df['getting_up'].replace(['0','1','2','3','4','not asked on pretest'], ['never in the past four weeks','less than a week','1 or 2 times a week','3 or 4 times a week','5 or more times a week','never in the past four weeks'], inplace=True)
df['getting_up'].value_counts()
```

```python
rp.summary_cat(df['getting_up'])
```

```python
#Graphs and tables or statistics
plt.xlabel("getting_up")
plt.ylabel("Percent")
plt.title("How often did you have trouble staying asleep throughout the night?")
df['getting_up'].value_counts(normalize = True).plot(kind ='bar', color = 'Orange', width = 0.2, alpha = 0.5)
```

```python
plt.title("How often did you have trouble staying asleep throughout the night?")
df['getting_up'].value_counts().plot(kind ='pie', autopct='%1.1f%%', colors=['purple', 'black', 'yellow', 'green'])
```

```python
#Manage data for variable 1 df['getting_up']= df['H4SP6']
df['things_in_past']= df['H4PE30']
df['things_in_past']=df['things_in_past'].loc[ df['things_in_past']!=' ' ]
df['things_in_past']=df['things_in_past'].loc[ df['things_in_past']!='6' ]
df['things_in_past']=df['things_in_past'].loc[ df['things_in_past']!='8' ]
df['things_in_past']=df['things_in_past'].loc[ df['things_in_past']!='.' ]
df['things_in_past'].value_counts()
```

```python
df['things_in_past'].replace(['1','2','3','4','5'], ['Strongly_Agree','Agree','Neither','Disagree','Strongly_Disagree'], inplace=True)
df['things_in_past'].value_counts()
```

```python
 rp.summary_cat(df['things_in_past'])
```

```python
#Graphs and tables or statistics
plt.xlabel("Not worrying of past")
plt.ylabel("Percent")
plt.title("I don't worry about things that have already happened")
df['things_in_past'].value_counts(normalize = True).plot(kind ='bar', color = 'purple', width = 0.2, alpha = 0.5)
```

```python
plt.title("I don't worry about things that have already happened")
df['things_in_past'].value_counts().plot(kind ='pie', autopct='%1.1f%%', colors=['olivedrab', 'rosybrown', 'gray', 'maroon'])
```
